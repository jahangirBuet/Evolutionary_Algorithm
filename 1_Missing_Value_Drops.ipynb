```python
df=df.drop(['EOD Primary Tumor',
'Lymphoma - Ann Arbor Stage',
'Grade Pathological',
'Summary stage',
'Derived EOD 2018 Stage Group',
'Grade Clinical',
'Lymph Node Size Recode'

],axis=1)
```

```python
for col in df.columns:
    if len(df[col].unique()) == 1:
        df.drop(col,inplace=True,axis=1)
```

```python
df.isnull().sum()
```

```python
def value_counts(x):
  print("                  "+x)
  print()
  print(df[x].value_counts())
  print()
columns=df.columns
for x in columns:
  value_counts(x)
```

```python
df=df.drop([
            'CS site-specific factor 25',
            'Primary by international rules',
            'First malignant primary indicator',
            'TNM 7/CS v0204+ Schema recode'



],axis=1)
```

```python
df1=df[ df['Vital status recode'] == 'Dead' 	]
```

```python
df1
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
df1=df1.dropna()
```

```python
df1.shape
```

```python
df2=df[ df['Vital status recode'] == 'Alive' 	]
```

```python
df2.isnull().sum()
```

```python
df2=df2.dropna()
```

```python
df2.shape
```

```python
df=pd.concat([df1, df2])
```
